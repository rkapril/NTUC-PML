Our Digits dataset is composed of 1797 samples of images, each image has 64 features - > 8x8 size images

```python
#Let's visualise a single digit
im1 = np.reshape(digits.data[0], (8,8))#reshape our data
plt.imshow(im1, cmap=plt.cm.gray_r)#show as a image
```

```python
n_samples = len(digits.images)
#convert our 8x8 images into a 64 length/size array for a training
data_images = digits.images.reshape((n_samples, -1))
```

#### Perform train test split

```python
from sklearn.model_selection import train_test_split

#split our data into 70% train, 30% test
X_train, X_test, y_train, y_test = train_test_split(data_images, digits.target, test_size=0.3)
```

#### Train our SVM model

```python
from sklearn import svm#Step 1: import SVM algorithm

svmclass = svm.SVC()#Step 2: Instatiate our model

svmclass.fit(X_train, y_train)#Step 3: Fit our model
print("Training is complete")
```

#### Get predictions

```python
y_pred = svmclass.predict(X_test)#Step 4: to get predictions
```

```python
#check our predictions
print("Original :", str(y_test[:5]))
print("Prediction :", str(y_pred[:5]))
```

#### Get Evaluation Metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy is :", str(accuracy_score(y_test, y_pred)))
```

```python
print("Confusion Matrix is :", str(confusion_matrix(y_test, y_pred)))
```
