### Logistic Regression

The last time we were looking linear regression, that it worked well for predicting continous outputs The idea is that linear regression does not fare well/work with classification problems.

If a problem is non-linear in nature, we will have to use logistic regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("import our libraries!")
```

```python
#titantic dataset

url = "http://bit.ly/wkspdata"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

### Data Pre-processing

```python
#drop some columns - cabin, name, ticket

dropcols = ["cabin", "name", "ticket"]
#drop the cols
#axis = 1 means cols, 0 means row
df = df.drop(dropcols, axis=1)
```

```python
df.info()
```

```python
#drop null values
df = df.dropna()
```

```python
df.info()
```

```python
#pclass, sex and embarked are categorical features => we need to convert them to numeric

cat_features = ["pclass", "sex", "embarked"]
df = pd.get_dummies(df, columns=cat_features)
df
```

```python
df.info()
```

#### Split our data for logistic regression

```python
from sklearn.model_selection import train_test_split
```

```python
#get our input and our output
#X and y

X = df.iloc[:, 1:]
y = df.iloc[:,0]
```

```python
X
```

```python
y
```

```python
#train 70% and test 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression#Step 1: import Logistic regression
log_reg = LogisticRegression(solver="newton-cg")#Step 2: initialise our model
log_reg.fit(X_train, y_train)#Step 3: train our model

print("Training complete!")
```

#### Prediction from model

```python
y_pred = log_reg.predict(X_test)#Step 4: obtain model predictions
print(y_pred)
```

```python
df1 = pd.DataFrame({"Original Survived":y_test, "Prediction" : y_pred})
df1
```

### Model Evaluation
We ae going to take a look at our performance metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

test_acc = accuracy_score(y_test, y_pred)#Step 5: model evaluation
print("Accuracy of :", str(test_acc))
```

```python
## Lets take a Look at confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
#precision score
test_precusion = precision_score(y_test, y_pred)
print("Accuracy of :", str(test_precusion))
```

```python
#recall score
test_recall = recall_score(y_test, y_pred)
print("Precision of :", str(test_recall))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

print("imported our libriaies!")
```

```python
#Load our iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

iris_target = iris.target #this is your y!
```

```python
iris_target #iris sentosa, iris versicolor, iris virginica
```

```python
idf = iris_df.copy()
idf["Iris_class"] = iris_target
idf.head()
```

```python
#process our data - split our data into train test split
from sklearn.model_selection import train_test_split

#train 70% and test 30%
X_train, X_test, y_train, y_test = train_test_split(iris_df, iris_target, test_size=0.3)
```

### Peform logsitic regression

```python
from sklearn.linear_model import LogisticRegression#Step 1: import Logistic regression
log_reg = LogisticRegression(solver="newton-cg")#Step 2: initialise our model
log_reg.fit(X_train, y_train)#Step 3: train our model

print("Training complete!")
```

```python
y_pred = log_reg.predict(X_test)#Step 4: obtain model predictions
print(y_pred)
```

```python
df1 = pd.DataFrame({"Original":y_test, "Prediction" : y_pred})
df1
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

test_acc = accuracy_score(y_test, y_pred)#Step 5: model evaluation
print("Accuracy of :", str(test_acc))
```

```python
## Lets take a Look at confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
#precision score
test_precusion = precision_score(y_test, y_pred, average=None)
print("Accuracy of :", str(test_precusion))
```

```python
#recall score
test_recall = recall_score(y_test, y_pred, average=None)
print("Precision of :", str(test_recall))
```
