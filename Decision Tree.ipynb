### Decision Trees

Let's create a decison tree from IRIS datasets!


```python
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
#target y
y = iris.target
```

```python
y
```

```python
#split our data, 70% train and 30% test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_df, y, test_size=0.3)
```

### Lets set up our decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier #Step 1: import our model
dtree = DecisionTreeClassifier()#Step 2: instantiate our model
dtree.fit(X_train, y_train)#Step 3: Train our model
print("Decision tree has been trained")
```

#### Get Our Predictions
> 0 - setosa
>
> 1 - Versicolor
>
> 2 - Virginica

```python
y_pred = dtree.predict(X_test) #Step 4: Get predictions
```

```python
print("Original :", str(y_test[:5]))
print("Prediction : ", str(y_pred[:5]))
```

#### Model Evaluation

```python
from sklearn.metrics import accuracy_score
#Step 5: Score our model
acc = accuracy_score(y_test, y_pred)
print("Accuracy is : ", str(acc))
```

### Let's Visualise our decision tree! We can view the rules which have been created

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(25,16))
atree = tree.plot_tree(dtree, feature_names=iris.feature_names, class_names=iris.target_names, 
                       filled=True, rounded=True, fontsize=14)
```
